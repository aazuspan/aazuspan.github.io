## Minified

```python
url = "https://code.earthengine.google.com/repo/file/load?repo=users%2Faazuspan%2Fminify_test&path=msslib_mini"


elapsed = []
total = []

for i in tqdm(range(100)):
    start = time.perf_counter()
    resp = requests.get(url, headers=headers)
    total_time = time.perf_counter() - start

    elapsed.append(resp.elapsed.total_seconds())
    total.append(total_time)
    
    time.sleep(0.25)

df_mini = pd.DataFrame({'server_s': elapsed, 'total_s': total})
df_mini["download_s"] = df_mini["total_s"] - df_mini["server_s"]
df_mini["version"] = "minified"
```

## Original

```python
url = "https://code.earthengine.google.com/repo/file/load?repo=users%2Faazuspan%2Fminify_test&path=msslib_original"

elapsed = []
total = []

for i in tqdm(range(100)):
    start = time.perf_counter()
    resp = requests.get(url, headers=headers)
    total_time = time.perf_counter() - start

    elapsed.append(resp.elapsed.total_seconds())
    total.append(total_time)
    
    time.sleep(0.25)

df_original = pd.DataFrame({'server_s': elapsed, 'total_s': total})
df_original["download_s"] = df_original["total_s"] - df_original["server_s"]
df_original["version"] = "original"
```
