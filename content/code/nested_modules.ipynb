```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
deep_times = [2.99, 2.88, 3.43, 3.38, 3.54, 3.06, 3.42, 3.35, 3.52, 2.80]
np.mean(deep_times)
```

```python
wide_times = [1.20, 1.27, 1.42, 1.35, 1.32, 1.19, 1.41, 1.45, 1.16, 1.24]
np.mean(wide_times)
```

```python
all_times = [0.653, 0.549, 0.566, 0.593, 0.556, 0.471, 0.500, 0.529, 0.758, 0.555]
np.mean(all_times)
```

```python
snazzy_times = [1.88, 1.95, 1.91, 1.98, 1.64, 3.89, 1.48, 1.68, 1.89, 2.03]
np.mean(snazzy_times)
```

```python
snazzy_mono_times = [0.505, 0.682, 0.590, 0.546, 0.479, 0.524, 0.537, 0.558, 0.538]
np.mean(snazzy_mono_times)
```

```python
df = pd.DataFrame({
    "chained": deep_times,
    "branched": wide_times,
    "monolith": all_times,
}).melt()
```

```python
fig = px.box(
    df, x="value", color="variable", 
    labels={"value": "Total Request Time (sec)", "variable": "Module Structure"}, 
    width=800, height=300, template="plotly_white")

fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.8,
    bgcolor="rgba(255, 255, 255, 0)",
), 
margin=dict(l=20, r=20, t=20, b=20),
legend_title_text="", font_size=18,
width=700, height=250
)

fig
```

```python
fig.write_html("../assets/nested_module_imports.html")
```
